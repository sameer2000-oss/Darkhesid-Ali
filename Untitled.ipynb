# Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = {
    "name": ["Riya", "Aman", "Priya", "Dev", "Sara"],
    "marks": [85, 78, 92, 55, 60],
    "subject": ["Math", "Science", "Math", "English", "Science"]
}
df = pd.DataFrame(data)

# Add grade column
def get_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "F"

df["grade"] = df["marks"].apply(get_grade)
print(df)


```

```python

# -------- Bar Chart: Names vs Marks --------
plt.figure(figsize=(6,4))
colors = df["grade"].map({"A":"green", "B":"blue", "C":"orange", "F":"red"})
plt.bar(df["name"], df["marks"], color=colors)
plt.title("Marks by Student")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()
```

```python
# -------- Histogram: All Marks --------
plt.figure(figsize=(6,4))
plt.hist(df["marks"], bins=5, color="skyblue", edgecolor="black")
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python

# -------- Line Chart: Average Marks per Subject --------
avg_marks = df.groupby("subject")["marks"].mean()
plt.figure(figsize=(6,4))
plt.plot(avg_marks.index, avg_marks.values, marker="o", linestyle="-", color="purple")
plt.title("Average Marks per Subject")
plt.xlabel("Subject")
plt.ylabel("Average Marks")
plt.show()
```

```python
# --- Step 6 (BONUS): Heatmap for Subject-wise Average Marks ---
grouped_df = df.groupby("subject")[["marks"]].mean().T  # transpose for better view
plt.figure(figsize=(5,2))
sns.heatmap(grouped_df, annot=True, cmap="coolwarm", fmt=".1f")
plt.title("Heatmap: Subject-wise Average Marks")
plt.show()
```
