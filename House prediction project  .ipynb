```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Step 1: Create dataset
data = {
    'size': [500, 600, 700, 800],
    'price': [300000, 350000, 400000, 450000]
}
df = pd.DataFrame(data)

# Step 2: Split data
X = df[["size"]]
y = df["price"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Step 3: Train model
model = LinearRegression()
model.fit(X, y)

# Step 4: Evaluate model
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("ðŸ“Š Mean Squared Error (MSE):", mse)
print("ðŸ“ˆ RÂ² Score:", r2)

# Step 5: Predict new value âœ…
new_sizes = pd.DataFrame([[500]], columns=["size"])
predicted_prices = model.predict(new_sizes)

for s, p in zip(new_sizes["size"], predicted_prices):
    print(f"Predicted price for {s} sq ft: â‚¹{p:.2f}")
```
