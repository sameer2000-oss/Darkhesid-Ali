# Loan approval

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# 1Ô∏è‚É£ Create simple loan dataset
data = {
    'income': [30000, 50000, 70000, 90000, 110000, 25000, 45000, 85000, 120000, 60000],
    'credit_score': [600, 650, 700, 720, 750, 580, 640, 710, 780, 680],
    'loan_amount': [300000, 200000, 250000, 150000, 200000, 350000, 300000, 180000, 220000, 260000],
    'approve': [0, 0, 1, 1, 1, 0, 0, 1, 1, 1]   # 1 = Approve, 0 = Reject
}

df = pd.DataFrame(data)
print("‚úÖ Sample Data:")
print(df)
```

```python
# 2Ô∏è‚É£ Separate features (X) and target (y)
X = df[['income', 'credit_score', 'loan_amount']]
y = df['approve']

# 3Ô∏è‚É£ Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

# 4Ô∏è‚É£ Create and train Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# 5Ô∏è‚É£ Test the model
accuracy = model.score(X_test, y_test)
print("\nüìä Model Accuracy:", round(accuracy, 2))
```

```python
# 6Ô∏è‚É£ Make predictions for new applicants
def predict_loan(income, credit_score, loan_amount):
    pred = model.predict(pd.DataFrame([[income, credit_score, loan_amount]], columns=["income","credit_score","loan_amount"]))
    result = pred[0]
    if result == 1:
        print(f"‚úÖ Loan Approved (Income: {income}, Credit Score: {credit_score}, Loan: {loan_amount})")
    else:
        print(f"‚ùå Loan Rejected (Income: {income}, Credit Score: {credit_score}, Loan: {loan_amount})")
```

```python
# 7Ô∏è‚É£ Example predictions
print("\nüè¶ Example Predictions:")
predict_loan(80000, 720, 200000)
predict_loan(30000, 600, 400000)
predict_loan(100000, 750, 250000)
```
